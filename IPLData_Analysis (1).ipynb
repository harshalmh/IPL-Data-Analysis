#IPL DATA ANALYSIS

```python
import pandas as pd
data=pd.read_csv("ipl_2008.csv")
```

```python
data
```

#No. of matches won by each team in IPL 2008

```python
import matplotlib.pyplot as plt
winner_counts = data["winner"].value_counts()

plt.bar(winner_counts.index, winner_counts.values)
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Number of Matches Won")
plt.title("number of matches won in IPL 2008")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data["won_by"] = data.apply(lambda row: "defending" if row["win_by_runs"] > 0 else "chasing", axis=1)

won_by_counts = data["won_by"].value_counts()


plt.figure(figsize=(6, 6))
plt.pie(won_by_counts, labels=won_by_counts.index, autopct="%1.1f%%", colors=['gold', 'lightgreen'] )
plt.title("Number of Matches Won by Defending or Chasing (IPL 2008)")
plt.show()
```

```python
last_10_matches = data.tail(10)
player_of_match_counts = last_10_matches["player_of_match"].value_counts()
plt.figure(figsize=(10, 5))
plt.bar(player_of_match_counts.index, player_of_match_counts.values, color='skyblue')
plt.xticks(rotation=45, fontsize=10)
plt.xlabel("Player")
plt.ylabel("Number of Awards")
plt.title("Player of the Match Awards (Last 10 Matches)")
plt.subplots_adjust(bottom=0.3)
plt.show()
```

```python
toss_winner_counts = data["toss_winner"].value_counts()
plt.figure(figsize=(12, 6))
plt.bar(toss_winner_counts.index, toss_winner_counts.values, color='lightcoral')
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("Team")
plt.ylabel("Number of Toss Wins")
plt.title("Maximum Toss Winner Team in IPL 2008")
plt.subplots_adjust(bottom=0.3)
plt.show()
```
